```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```

```python
df.info()
```

```python
df['Year'] = df['Timestamp'].dt.year
df['Month'] = df['Timestamp'].dt.month_name()
df['Quarter'] = df['Timestamp'].dt.quarter
df['Week'] = df['Timestamp'].dt.isocalendar().week
```

```python
df['Like_Rate'] = (df['Likes'] / df['Views']) * 100
df['Comment_Rate'] = (df['Comments'] / df['Views']) * 100
df['Share_Rate'] = (df['Shares'] / df['Views']) * 100
df['Save_Rate'] = (df['Saves'] / df['Views']) * 100
```

```python
df['Engagement_Score'] = (
    df['Likes'] +
    df['Comments'] +
    df['Shares'] +
    df['Saves']
)
```

```python
df.head()
```

```python
df.to_csv("cleaned_social_media.csv", index=False)

print("Cleaned dataset saved successfully!")
```

```python
# Number of posts by platform
platform_counts = df['Platform'].value_counts()

print(platform_counts)

# Bar Chart
platform_counts.plot(kind='bar', figsize=(8,5))

plt.title("Number of Posts by Platform")
plt.xlabel("Platform")
plt.ylabel("Number of Posts")
plt.xticks(rotation=45)
plt.show()
```

```python
content_counts = df['Content_Type'].value_counts()

print(content_counts)

content_counts.plot(kind='bar', figsize=(8,5))

plt.title("Content Type Distribution")
plt.xlabel("Content Type")
plt.ylabel("Number of Posts")
plt.xticks(rotation=45)
plt.show()
```

```python
category_counts = df['Category'].value_counts()

print(category_counts)

category_counts.plot(kind='bar', figsize=(10,5))

plt.title("Posts by Category")
plt.xlabel("Category")
plt.ylabel("Number of Posts")
plt.xticks(rotation=45)
plt.show()
```

```python
avg_likes = df.groupby('Platform')['Likes'].mean().sort_values(ascending=False)

print(avg_likes)

avg_likes.plot(kind='bar', figsize=(8,5))

plt.title("Average Likes by Platform")
plt.xlabel("Platform")
plt.ylabel("Average Likes")
plt.xticks(rotation=45)
plt.show()
```

```python
avg_engagement = df.groupby('Platform')['Engagement_Rate'].mean().sort_values(ascending=False)

print(avg_engagement)

avg_engagement.plot(kind='bar', figsize=(8,5))

plt.title("Average Engagement Rate by Platform")
plt.xlabel("Platform")
plt.ylabel("Engagement Rate")
plt.xticks(rotation=45)
plt.show()
```

```python
top_posts = df.nlargest(10, 'Engagement_Score')

print(top_posts[['Post_ID','Platform','Content_Type','Engagement_Score']])
```

```python
best_day = df.groupby('Day_of_Week')['Engagement_Score'].mean()

print(best_day)

best_day.plot(kind='bar', figsize=(8,5))

plt.title("Average Engagement by Day")
plt.xlabel("Day")
plt.ylabel("Average Engagement Score")
plt.xticks(rotation=45)
plt.show()
```

```python
best_hour = df.groupby('Hour_of_Day')['Engagement_Score'].mean()

print(best_hour)

best_hour.plot(figsize=(10,5))

plt.title("Average Engagement by Posting Hour")
plt.xlabel("Hour")
plt.ylabel("Average Engagement Score")
plt.grid(True)

plt.show()
```

```python
tier = df.groupby('Influencer_Tier')['Engagement_Score'].mean()

print(tier)

tier.plot(kind='bar', figsize=(8,5))

plt.title("Engagement by Influencer Tier")
plt.xlabel("Influencer Tier")
plt.ylabel("Average Engagement Score")

plt.show()
```

```python
sentiment = df.groupby('Sentiment')['Engagement_Score'].mean()

print(sentiment)

sentiment.plot(kind='bar', figsize=(8,5))

plt.title("Average Engagement by Sentiment")
plt.xlabel("Sentiment")
plt.ylabel("Average Engagement Score")

plt.show()
```

```python
monthly = df.groupby('Month')['Engagement_Score'].mean()

print(monthly)

monthly.plot(figsize=(10,5))

plt.title("Monthly Engagement Trend")
plt.xlabel("Month")
plt.ylabel("Average Engagement Score")
plt.grid(True)

plt.show()
```

```python
correlation = df[['Likes','Comments','Shares','Views','Saves','Follower_Count','Engagement_Score']].corr()

print(correlation)
```
